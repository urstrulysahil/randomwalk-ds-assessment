```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the dataset
books_df = pd.read_csv('books.csv')
```

```python
# Q1: How many rows and columns are there in books.csv dataset?

rows, columns = books_df.shape
print(f"Q1: Rows = {rows}, Columns = {columns}")
```

```python
# Q2: How many books do not have an original title?

books_without_original_title = books_df['original_title'].isnull().sum()
print(f"Q2: Books without original title = {books_without_original_title}")
```

```python
# Q3: How many unique books are present in the dataset ? Evaluate based on the 'book_id' after removing null values in the original_title column.

unique_books = books_df['book_id'].nunique()
print(f"Q3: Unique books = {unique_books}")
```

```python
# Q4: What is the average rating of all the books in the dataset based on ‘average_rating’?

average_rating = books_df['average_rating'].mean()
print(f"Q4: Average rating of all books = {average_rating:.2f}")
```

```python
# Q5: Find the number of books published in the year ‘2000’ based on the ‘original_publication_year’.

books_published_in_2000 = books_df[books_df['original_publication_year'] == 2000].shape[0]
print(f"Q5: Books published in 2000 = {books_published_in_2000}")
```

```python
# Q6: Which book (title) has the maximum number of ratings based on ‘work_ratings_count’.

max_rated_book = books_df.loc[books_df['work_ratings_count'].idxmax()]['title']
print(f"Q6: Book with maximum ratings = {max_rated_book}")
```

```python
# Q7: Bucket the average_rating of books into 11 buckets [0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0] with 0.5 decimal rounding (eg: average_rating 3.0 to 3.49 will fall in bucket 3.0). Plot bar graph to show total number of books in each rating bucket.

rating_buckets = [0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0]
books_df['rating_bucket'] = pd.cut(books_df['average_rating'], bins=rating_buckets, include_lowest=True)
rating_counts = books_df['rating_bucket'].value_counts().sort_index()
# Plotting the bar graph
rating_counts.plot(kind='bar', edgecolor='black')
plt.title('Number of Books in Each Rating Bucket')
plt.xlabel('Rating Bucket')
plt.ylabel('Number of Books')
plt.show()
```
